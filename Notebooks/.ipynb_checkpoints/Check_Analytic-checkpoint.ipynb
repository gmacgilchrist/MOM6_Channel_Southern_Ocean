# Skew Sinusoidal 

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, num=100)
```

```python
# sinusoidal
y = np.sin( x + np.sin(x)*0.5/2.)
y1 = np.sin( x + np.sin(2*x)*0.5/2.)
#y1 = 7.*np.sin(x)/16. + 7.*np.sin(2*x)/64. + 1.*np.sin(3*x)/48. + 1.*np.sin(4*x)/512.
#plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x, np.sin(x),'--')
plt.grid()
```

```python
# sinusoidal square
y = np.sin( x + np.sin(x)/2.)**2
y1 = np.sin( x - np.sin(x)/2.)**2
plt.plot(x,y)
plt.plot(x,y1,'--')
plt.plot(x, np.sin(x)**2,':')
plt.grid()
```
